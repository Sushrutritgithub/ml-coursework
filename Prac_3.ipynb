```python
df_clean = df.drop(columns=['Unnamed: 0'])
transactions = []
for i, row in df_clean.iterrows():
    basket = [item for item in df_clean.columns if row[item] == True]
    
    if basket:
        transactions.append(set(basket))
```

```python
print(transactions[:5])
```

```python
def get_support(transactions, itemset):
    cnt = 0
    for basket in transactions:
        if itemset.issubset(basket):
            cnt += 1
    return cnt / len(transactions)
```

```python
frequent_itemsets = apriori(df_clean, min_support=0.1, use_colnames=True)  
```

```python
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.2)
```
